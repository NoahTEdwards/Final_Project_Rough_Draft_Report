### Importing

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### The Data

Here I am loading in the csv file that contains all of the data for the brooklyn nets draft years.

```python
bn_draft = pd.read_csv(r'C:\Users\ntedw\EMAT 22110\Brooklyn_Nets_Draft.csv')
bn_draft.keys()
```

I am loading in all of the data.

```python
bn_draft.head(50)
```

Here I am creating Years and PPG to combine them together and make one column.

```python
Years = bn_draft['Year']
Years.head()
```

```python
PPG = bn_draft['PTS Per Game']
PPG.head()
```

Here I combine Years and PPG to create Years_PPG`

### Visualization 

In my visualization you will be able to two charts. One chart being a chart with the years in descending order and the other chart being a chart with the Points Per Game in descending order.

You will also be able to see two bar graphs. One graph with the years in descending order and one graph with the Points Per Game in descending order.

This chart is showing the years in descending order.

```python
Years_PPG = bn_draft[['Year','PTS Per Game']]
Years_PPG.head(40)
```

This graph is showing The years in descending order

```python
Years_PPG = Years_PPG.sort_values('Year', ascending = False)[:75]
Years_PPG

fig, ax0 = plt.subplots(figsize=(15,10))
ax0.bar(Years.index, Years_PPG['PTS Per Game'])

plt.xticks(rotation=90, fontsize = 'small')
plt.xlabel('Years')
plt.ylabel('Points Per Game')
plt.show()
```

This chart is chart is showing the Points Per Game in descending order.

```python
Years_PPG = Years_PPG.sort_values('PTS Per Game', ascending = False)[:75]
Years_PPG
```

This chart is showing the Points Per Game in descending order.

```python
fig, ax0 = plt.subplots(figsize=(15,10))
ax0.bar(Years.index, Years_PPG['PTS Per Game'])

plt.xticks(rotation=90, fontsize = 'small')
plt.xlabel('Years')
plt.ylabel('Points Per Game')
plt.show()
```

# Conclusion

In conclusion we can see that 2008 was the best draft year for the Brooklyn nets when it comes to Points Per Game average. If you look at the charts and graphs you can see the players drafted in 2008 had the best average for Points Per Game.

My hypothesis was correct. The best draft year for Points Per Game was a year in between 2000 - 2010.
